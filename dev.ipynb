# List ALL Intents

```python
def list_intents():
    # Create a client
    client = dialogflowcx_v3.IntentsClient()

    # Initialize request argument(s)
    request = dialogflowcx_v3.ListIntentsRequest(
        parent=parent_value,
    )

    # Make the request
    page_result = client.list_intents(request=request)

    # Handle the response
    for response in page_result:
        print(response)
# list_intents()
```

```python

cap = print(list_intents())
with open('intents', 'w') as f:
    f.write(cap.stdout)

print(cap)
```
